## 1) quantas pessoas estão de ferias ?

```python
print(f"Há {num_ferias} funcionários de férias atualmente.")
```

```python
## Gráfico da quantidade de pessoas que estão de ferias 
plt.barh(["Funcionários em férias"], [num_ferias])

# Configurações de Barras Horizontal 
plt.title("Número de funcionários de férias")
plt.xlabel("Quantidade")
plt.xlim(0, num_ferias + 1) 

# exibe o gráfico
plt.show()
```

## 2) Quantas pessoas estão programadas a cada mes por setor? 
## Operaçoes - Governança - Delivery - Soluçoes - arquitetura.

```python
# conversão da coluna data de início para uma nova coluna "MES"
ferias["MES"] = ferias["DATA_INICIO"].dt.month
```

```python
# agrupamento dos dados por mês e equipe que calcula o número total de pessoas programadas
total_por_mes_e_setor = ferias.groupby(["MES", "EQUIPE"]).size().reset_index(name="quantidade")
```

```python
## Gráfico do número de pessoas programadas para cada mês de equipe
fig, ax = plt.subplots(figsize=(10,6)) 
cores = ["#2c3e50", "#3498db", "#95a5a6", "#9b59b6", "#f1c40f"] 
for i, equipe in enumerate(total_por_mes_e_setor["EQUIPE"].unique()): 
    
    dados_equipe = total_por_mes_e_setor[total_por_mes_e_setor["EQUIPE"] == equipe] 
    ax.bar(dados_equipe["MES"], dados_equipe["quantidade"], color=cores[i], label=equipe) 
    
## Legendas e Titulos 
ax.legend() 
ax.set_xlabel("Mês") 
ax.set_ylabel("Número de pessoas programadas") 
ax.set_title("Programação de férias por mês e equipe") 
plt.show()
```

## 3) se tem conflito na programaçao das datas? e quais os nomes estao chocando as ferias.

```python
# conversão das colunas de data para o formato datetime
ferias["DATA_INICIO"] = pd.to_datetime(ferias["DATA_INICIO"])
ferias["DATA_FIM"] = pd.to_datetime(ferias["DATA_FIM"])
```

```python
## ordena os dados por nome e data de inicio 
ferias = ferias.sort_values(["MOME", "DATA_INICIO"]) 
```

```python
# Identificação de sobreposição entre as datas de inicio e fim de cada periodo de férias
ferias["CONFLITO"] = ((ferias["DATA_INICIO"] - ferias["DATA_FIM"].shift(1)).dt.days < 0) 
```

```python
# exibição dos funcionários que tem conflitos nas férias 
funcionarios_com_conflitos = ferias[ferias["CONFLITO"]]["MOME"].unique() 

if len(funcionarios_com_conflitos) > 0: 
    
    print("Os seguintes funcionários têm conflitos de férias: ") 
    
    for funcionario in funcionarios_com_conflitos: 
        
        print(funcionario) 
        
else: 
    
    print("Não há conflitos de férias.") 
```

```python
## Conta o número de conflitos de férias para cada funcionário 
conflitos_por_funcionario = ferias[ferias["CONFLITO"]]["MOME"].value_counts()
```

```python
# Cria um gráfico de barras
plt.scatter(x=conflitos_por_funcionario.index, y=conflitos_por_funcionario)
plt.xticks(rotation=90) 
plt.title("Número de conflitos de férias por funcionário") 
plt.xlabel("Funcionário") 
plt.ylabel("Número de conflitos") 
plt.show()
```
